```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

```

```python

group1_name = 'Pre-PTX Control' 
group2_name = 'PTX'
group3_name = 'Pre-CNQX Control'
group4_name = 'CNQX'

# Data for the bar graph = []
group1 = [16.13766751,	17.16545952,	13.18101239,	25.18307182,	21.40734783]
group2 = [13.09847251,	15.02006699,	9.425121674,	18.17339634,	18.30409854]
group3 = [21.9802961,	28.06239495,	22.8304807,	16.65480202]
group4 = [27.41541837,	41.56042628,	25.73492274,	21.20759404]

# Combine data
vals = group1 + group2 + group3 + group4
GROUPS = [group1_name]*len(group1) + [group2_name]*len(group2) + [group3_name]*len(group3) + [group4_name]*len(group4)
```

```python
# Create the bar graph
plt.figure(figsize=(4,4))

ax = sns.barplot(x=GROUPS, y=vals, capsize=0.3, edgecolor='0.2', linewidth=2.5, 
                 hue=GROUPS, palette=['brown', 'mistyrose', 'blue', 'lightblue'], legend=False,
                 err_kws={'color': '0.2', 'linewidth': 2.5})

kwargs = {'edgecolor': '0.2', 'linewidth': 2.5, 'fc': 'WHITE'}
ax = sns.swarmplot(x=GROUPS, y=vals, marker='s', s=6, **kwargs)

# Add significance markers
max_val = max(vals)
plt.plot([0,0,1,1], [max_val+3,max_val+4,max_val+4,max_val+3], lw=2.5, color='0.2')
plt.text(x=0.5, y=max_val+4, s='*', ha='center', size=20, weight='bold', color='0.2')

#plt.plot([2,2,3,3], [max_val+3,max_val+4,max_val+4,max_val+3], lw=2.5, color='0.2')
#plt.text(x=2.5, y=max_val+4, s='*', ha='center', size=20, weight='bold', color='0.2')

# Customize the graph
ax.set_ylabel('OBR (dB)', fontsize=14, weight='bold', color='0.2', rotation=90)

for axis in ['bottom', 'left']:
    ax.spines[axis].set_linewidth(2.5)
    ax.spines[axis].set_color('0.2')

plt.xticks(size=14, rotation=35, rotation_mode='anchor', ha='right', weight='bold', color='0.2')
plt.yticks(size=14, weight='bold', color='0.2')

ax.tick_params(width=2.5, color='0.2')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
