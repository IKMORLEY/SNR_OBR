```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

```

```python

group1_name = 'PTX' 
group2_name = 'CNQX'

# Data for the bar graph = []
group1 = [-0.01,	0.19,	0.19,	0.11,	-0.03]
group2 = [-0.05,	0.02,	-0.07,	-0.07]

# Combine data
vals = group1 + group2
GROUPS = [group1_name]*len(group1) + [group2_name]*len(group2) 
```

```python
plt.figure(figsize=(3,6))

ax = sns.barplot(x=GROUPS, y=vals, capsize=0.3, edgecolor='0.2', linewidth=2.5, 
                 hue=GROUPS, palette=['brown', 'aqua'], legend=False,
                 err_kws={'color': '0.2', 'linewidth': 2.5})

kwargs = {'edgecolor': '0.2', 'linewidth': 2.5, 'fc': 'WHITE'}
ax = sns.swarmplot(x=GROUPS, y=vals, marker='s', s=6, **kwargs)

#significance or not 
max_val = max(vals)
min_val = min(vals)
#plt.text(x=0.0, y=min_val-3, s='*', ha='center', size=20, weight='bold', color='0.2')
# plt.text(x=1.0, y=max_val+4, s='*', ha='center', size=20, weight='bold', color='0.2')

ax.set_ylabel('Mean Change in EPSP Amplitude (mV)', fontsize=14, weight='bold', color='0.2', rotation=90)

for axis in ['bottom', 'left']:
    ax.spines[axis].set_linewidth(2.5)
    ax.spines[axis].set_color('0.2')

plt.xticks(size=14, rotation=35, rotation_mode='anchor', ha='right', weight='bold', color='0.2')
plt.yticks(size=14, weight='bold', color='0.2')

ax.tick_params(width=2.5, color='0.2')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set y-axis range if super negative values
#plt.ylim(-12, max_val+1)

# Adjust layout to prevent cutting off labels
plt.tight_layout()
```
