```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data
x = np.array([78.2,	138,	60.4,	44,	17.04,	21.28,	9.6,	52.6,	36.56])
y = np.array([23.792,	58.376,	21.952,	22.404,	6.016,	10.892,	5.5,	25.2,	36.172])

# Perform linear regression to calculate the best fit line and R value
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Calculate the best fit line
y_fit = slope * x + intercept

# Plot the data points and the best fit line
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(x, y, marker='s', facecolors='blue', edgecolors='0.2', linewidth=2.5, label='Data Points')  # Hollow squares
ax.plot(x, y_fit, color='0.2', linewidth=2.5, label=f'Best Fit Line (R={r_value:.2f})')  # Best fit line in color '0.2'

# Set axis labels and title with the specified style
ax.set_xlabel('Mean AP Frequency', fontsize=14, weight='bold', color='0.2')
ax.set_ylabel('Mean Power (yW)', fontsize=14, weight='bold', color='0.2')
ax.set_title('AP Frequency Correlates with Power', fontsize=16, weight='bold', color='0.2')

# Customize legend
ax.legend()

# Customize grid
ax.grid(True, color='0.8', linestyle='--', linewidth=1)

# Customize tick parameters
ax.tick_params(axis='both', which='major', labelsize=12, colors='0.2')

# Remove upper and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Style remaining spines (left and bottom)
ax.spines['left'].set_color('0.2')
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_color('0.2')
ax.spines['bottom'].set_linewidth(2)

plt.show()
```
